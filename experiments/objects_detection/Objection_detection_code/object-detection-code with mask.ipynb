```python
import matplotlib.pyplot as plt
import cvlib as cv
from cvlib.object_detection import draw_bbox
import random

data_path = '/kaggle/input/object-detection/object_detection/'

img_name = random.choice(os.listdir(data_path))
im = cv2.imread(data_path + img_name)
im = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
bbox, label, conf = cv.detect_common_objects(im)
output_image = draw_bbox(im, bbox, label, conf)
plt.figure(figsize=(15,15))
plt.imshow(output_image)
plt.show()
```

```python
 for box, lbl, cnf in zip(bbox, label, conf):
        print(f'{lbl} is in {box} with {cnf:.0%} confidence')
```
